```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
customers = pd.read_csv(r'Datas/Customers.csv', encoding='latin1')
```

```python
customers
```

```python
def convert_date(date_str):
    try:
        return pd.to_datetime(date_str, errors='coerce').strftime('%Y-%m-%d')
    except:
        return date_str 
customers['Birthday'] = customers['Birthday'].apply(convert_date)
```

```python
customers
```

```python
customers.isnull()
```

```python
customers.isnull().sum()
```

```python
customers.isnull().sum().sum()
```

```python
customers.dropna()
```

```python
customers = customers.assign(**{'StateCode': customers['StateCode'].fillna('Unavailable')})
```

```python
customers.isnull()
```

```python
customers.isnull().sum()
```

```python
customers.isnull().sum().sum()
```

```python
customers
```

```python
customers.to_csv('Customers.csv', index=False)
```

```python
def parse_currency(value):
    try:
        return float(value.replace('$', ''))
    except ValueError:
        return None

products['UnitCostUSD'] = products['UnitCostUSD'].apply(parse_currency)
products['UnitPriceUSD'] = products['UnitPriceUSD'].apply(parse_currency)
```

```python
products
```

```python
products.to_csv('ConvertedDatas\products.csv', index=False)
```

```python
import sqlite3

conn = sqlite3.connect('Global_Electronics.db')
cursor = conn.cursor()

csv_file = 'ConvertedDatas\Customers.csv' 
df = pd.read_csv(csv_file, encoding='ISO-8859-1')

df.to_sql('customers', conn, if_exists='replace', index=False)


csv_file = 'ConvertedDatas\Exchange_Rates.csv' 
df = pd.read_csv(csv_file, encoding='ISO-8859-1')

df.to_sql('exchange', conn, if_exists='replace', index=False)


csv_file = 'ConvertedDatas\Products.csv' 
df = pd.read_csv(csv_file, encoding='ISO-8859-1')

df.to_sql('products', conn, if_exists='replace', index=False)

csv_file = 'ConvertedDatas\Sales.csv' 
df = pd.read_csv(csv_file, encoding='ISO-8859-1')
df.to_sql('sales', conn, if_exists='replace', index=False)

csv_file = 'ConvertedDatas\Stores.csv' 
df = pd.read_csv(csv_file, encoding='ISO-8859-1')
df.to_sql('stores', conn, if_exists='replace', index=False)


conn.close()
```
